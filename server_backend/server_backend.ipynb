```python
%%writefile rag.py
from typing import Any, List, Optional
from fastapi import FastAPI, HTTPException, UploadFile, File, Form
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import torch
import tensorflow as tf
import os
import uuid
from contextlib import asynccontextmanager


from typing import List, Tuple, Dict
import hashlib
import numpy as np
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from rank_bm25 import BM25Okapi
from langchain.prompts.chat import ChatPromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import pipeline
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import fitz  # PyMuPDF
from langchain.document_loaders import PyPDFLoader
from transformers import BitsAndBytesConfig
from langchain.text_splitter import TokenTextSplitter
from langchain_community.vectorstores.utils import DistanceStrategy
import re
from sentence_transformers import SentenceTransformer, util
import uuid
import json
import os

class ContextualRetrieval:

  def __init__(self, conversation_memory_size=5):

    self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    self.model_name_RAG = "AITeamVN/Vi-Qwen2-7B-RAG"

    self.quantization_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_compute_dtype=torch.float16,  # Thay đổi từ bfloat16 sang float16
        bnb_4bit_quant_type="nf4",             # Thêm loại lượng tử hóa tối ưu
        llm_int8_skip_modules=["lm_head"],     # Bỏ qua lượng tử hóa cho module này
        llm_int8_threshold=6.0,                # Ngưỡng tối ưu
        bnb_4bit_quant_storage=torch.uint8     # Sử dụng lưu trữ uint8 cho hiệu suất
    )

    self.embeddings = HuggingFaceEmbeddings(model_name = "bkai-foundation-models/vietnamese-bi-encoder")

    self.tokenizer_RAG = AutoTokenizer.from_pretrained(self.model_name_RAG)

    self.model_RAG = AutoModelForCausalLM.from_pretrained(
      self.model_name_RAG,
      device_map="auto",
      quantization_config=self.quantization_config
    )


    self.conversation_history = []
    self.conversation_memory_size = conversation_memory_size


    self.current_session_id = None
    self.sessions = {}

    if torch.cuda.is_available():
      try:
        self.reranker = SentenceTransformer('intfloat/multilingual-e5-large')
        self.reranker.to(self.device)
        self.has_reranker = True
      except:
        self.has_reranker = False
    else:
      self.has_reranker = False

  def start_new_session(self):
    self.current_session_id = str(uuid.uuid4())
    self.sessions[self.current_session_id] = []
    return self.current_session_id

  def load_session(self, session_id):
    if session_id in self.sessions:
      self.current_session_id = session_id
      self.conversation_history = self.sessions[session_id]
      return True
    return False

  def add_to_conversation(self, query, response):
    if self.current_session_id is None:
      self.start_new_session()

    conv_pair = {"query": query, "response": response, "timestamp": torch.cuda.Event()}

    self.sessions[self.current_session_id].append(conv_pair)

    self.conversation_history = self.sessions[self.current_session_id][-self.conversation_memory_size:]

  def load_pdf(self, pdf_path):
      loader = PyPDFLoader(pdf_path)
      pages = loader.load()
      text_splitter = TokenTextSplitter(chunk_size=256, chunk_overlap=48)
      docs = text_splitter.split_documents(pages)

      return docs

  def embedding_faiss(self, docs, encoder = None):
      if encoder is None:
        encoder = self.embeddings

      faiss_db = FAISS.from_documents(docs, encoder, distance_strategy=DistanceStrategy.DOT_PRODUCT)

      return faiss_db

  def compress_documents(self, docs, compression_ratio=0.7):
    compressed_docs = []

    for doc in docs:
      text = doc.page_content

      text = re.sub(r'\s+', ' ', text)
      text = re.sub(r'\n+', '\n', text)

      sentences = text.split('.')

      if len(sentences) <= 3:
        compressed_text = text
      else:

        num_sentences = max(3, int(len(sentences) * compression_ratio))

        selected_sentences = [sentences[0]]

        middle_sentences = sentences[1:-1]
        if len(middle_sentences) > 0:
          step = max(1, len(middle_sentences) // (num_sentences - 2))
          selected_sentences.extend(middle_sentences[::step][:num_sentences-2])

        if len(sentences) > 1:
          selected_sentences.append(sentences[-1])

        compressed_text = '.'.join(selected_sentences)

      compressed_doc = Document(
        page_content=compressed_text,
        metadata=doc.metadata
      )
      compressed_docs.append(compressed_doc)

    return compressed_docs

  def filter_documents(self, docs, query, threshold=0.3):
    query_embedding = self.embeddings.embed_query(query)

    filtered_docs = []
    for doc in docs:
      doc_embedding = self.embeddings.embed_query(doc.page_content)

      similarity = np.dot(query_embedding, doc_embedding) / (
        np.linalg.norm(query_embedding) * np.linalg.norm(doc_embedding)
      )

      if similarity >= threshold:
        filtered_docs.append(doc)

    if not filtered_docs and docs:
      filtered_docs = [docs[0]]

    return filtered_docs

  def rerank_documents(self, docs, query, top_k=5):
    if not self.has_reranker or not docs:
      return docs[:top_k] if len(docs) > top_k else docs

    reranker_query = f"query: {query}"

    passages = [doc.page_content for doc in docs]

    query_embedding = self.reranker.encode([reranker_query], convert_to_tensor=True)
    passage_embeddings = self.reranker.encode(passages, convert_to_tensor=True)

    scores = util.cos_sim(query_embedding, passage_embeddings)[0]

    doc_score_pairs = list(zip(docs, scores.cpu().numpy()))
    ranked_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True)

    reranked_docs = [pair[0] for pair in ranked_pairs[:top_k]]

    return reranked_docs

  def search_faiss(self, query, faiss_db, k=10):
    initial_k = min(k * 3, 20)  # Lấy số lượng document ban đầu lớn hơn
    retrieved_docs = faiss_db.similarity_search(query, k=initial_k)


    filtered_docs = self.filter_documents(retrieved_docs, query)


    reranked_docs = self.rerank_documents(filtered_docs, query, top_k=k)

    context = "".join(doc.page_content + "\n" for doc in reranked_docs)

    return context

  def generate_conversation_context(self):
    if not self.conversation_history:
      return ""

    context = "Lịch sử trò chuyện gần đây:\n"
    for i, conv in enumerate(self.conversation_history):
      context += f"Người dùng: {conv['query']}\n"
      context += f"Trợ lý: {conv['response']}\n"

    return context

  def enrich_query_with_context(self, query):
    # Phân tích query để xác định các tham chiếu đến thông tin trước đó
    has_reference = any(ref in query.lower() for ref in ["vừa nói", "như đã đề cập", "trước đó", "điều này", "điều đó", "vấn đề đó"])

    if len(query.split()) < 5 or has_reference:
      # Tìm các query gần đây có thể liên quan
      related_context = ""
      for conv in reversed(self.conversation_history):
        if any(word in conv['query'].lower() for word in query.lower().split()):
          related_context += f"Bạn đã hỏi: {conv['query']}\n"
          related_context += f"Trả lời: {conv['response']}\n"

      if related_context:
        return query + " (Tham khảo ngữ cảnh: " + related_context + ")"

    return query

  def inference(self, query: str, context: str, model = None, tokenizer = None):
      if model is None:
          model = self.model_RAG
      if tokenizer is None:
          tokenizer = self.tokenizer_RAG

      conversation_context = self.generate_conversation_context()

      if context == None or context == "":
          prompt = f"""Hãy cho câu trả lời chi tiết cho câu hỏi trên.
          {conversation_context}
          Câu hỏi: {query}"""
      else:
          prompt = f"""Sử dụng thông tin trong ngữ cảnh và lịch sử trò chuyện, hãy trả lời câu hỏi, lưu ý rằng phải trả lời bằng tiếng việt và trả lời dưới dạng một đoạn văn và không được thêm thắt gì.
              Ngữ cảnh tài liệu: {context}
              {conversation_context}
              Câu hỏi: {query}"""

      chat = [
          {"role": "user", "content": prompt},
      ]
      formatted_prompt = tokenizer.apply_chat_template(
          chat,
          tokenize=False,
          add_generation_prompt=True,
      )
      inputs = tokenizer.encode(
          formatted_prompt, add_special_tokens=False, return_tensors="pt"
      ).to(self.device)
      with torch.no_grad():
          outputs = model.generate(
              input_ids=inputs,
              max_new_tokens=250,
              do_sample=False,
          )
      full_response = tokenizer.decode(outputs[0], skip_special_tokens=False)

      # Xử lý phản hồi để loại bỏ các tag không mong muốn
      # Tìm và trích xuất nội dung giữa <|im_start|>assistant và <|im_end|>
      import re
      assistant_pattern = r'<\|im_start\|>assistant\n(.*?)<\|im_end\|>'
      matches = re.findall(assistant_pattern, full_response, re.DOTALL)

      if matches and len(matches) > 0:
          # Lấy phản hồi cuối cùng nếu có nhiều kết quả trùng khớp
          clean_response = matches[-1].strip()
      else:
          # Nếu không tìm thấy pattern, thử cách xử lý thông thường
          clean_response = full_response.replace(formatted_prompt, "", 1).strip()
          clean_response = re.sub(r'<\|im_start\|>.*?<\|im_end\|>', '', clean_response, flags=re.DOTALL)
          clean_response = clean_response.replace("<|im_start|>", "").replace("<|im_end|>", "")
          clean_response = clean_response.replace("<eos>", "").strip()

      self.add_to_conversation(query, clean_response)

      return clean_response
  def qa_chat(self, query, faiss_db):
    enriched_query = self.enrich_query_with_context(query)

    context = self.search_faiss(enriched_query, faiss_db)

    answer = self.inference(query, context)

    return answer

  def analyze_query_intent(self, query):
    # Phát hiện các mẫu câu chỉ dẫn ý định tiếp tục/thay đổi chủ đề
    continue_patterns = ["tiếp tục", "thêm về", "nói thêm", "giải thích thêm"]
    new_topic_patterns = ["chủ đề mới", "câu hỏi khác", "nói về", "giờ hỏi về"]


    is_continuation = any(pattern in query.lower() for pattern in continue_patterns)
    is_new_topic = any(pattern in query.lower() for pattern in new_topic_patterns)

    return {
      "is_continuation": is_continuation,
      "is_new_topic": is_new_topic
    }

  def contextual_qa_chat(self, query, faiss_db):
    intent = self.analyze_query_intent(query)
    if intent["is_continuation"] and self.conversation_history:
      last_query = self.conversation_history[-1]["query"]
      combined_query = last_query + " " + query
      context = self.search_faiss(combined_query, faiss_db)
    elif intent["is_new_topic"]:
      context = self.search_faiss(query, faiss_db)
    else:
      enriched_query = self.enrich_query_with_context(query)
      context = self.search_faiss(enriched_query, faiss_db)

    answer = self.inference(query, context)

    return answer

  def save_conversation_history(self, file_path):

    history_data = {}
    for session_id, conversations in self.sessions.items():
      history_data[session_id] = [
        {"query": c["query"], "response": c["response"]}
        for c in conversations
      ]

    with open(file_path, 'w', encoding='utf-8') as f:
      json.dump(history_data, f, ensure_ascii=False, indent=2)

  def load_conversation_history(self, file_path):


    if not os.path.exists(file_path):
      return False

    with open(file_path, 'r', encoding='utf-8') as f:
      history_data = json.load(f)

    for session_id, conversations in history_data.items():
      self.sessions[session_id] = [
        {"query": c["query"], "response": c["response"], "timestamp": torch.cuda.Event()}
        for c in conversations
      ]

    return True

  def process_documents(self, pdf_path, use_compression=True):
    """Xử lý tài liệu với nén dữ liệu và tạo vector store"""
    # Tải tài liệu
    docs = self.load_pdf(pdf_path)

    # Nén tài liệu nếu được yêu cầu
    if use_compression:
      docs = self.compress_documents(docs)

    # Tạo vector store
    faiss_db = self.embedding_faiss(docs)

    return faiss_db
```

```python
%%writefile app.py
from typing import Any, List, Optional
from fastapi import FastAPI, HTTPException, UploadFile, File, Form
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import torch
import tensorflow as tf
import os
import uuid
from contextlib import asynccontextmanager
from pymongo import MongoClient

from typing import List, Tuple, Dict
import hashlib
import numpy as np
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from rank_bm25 import BM25Okapi
from langchain.prompts.chat import ChatPromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import pipeline
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import fitz  # PyMuPDF
from langchain.document_loaders import PyPDFLoader
from transformers import BitsAndBytesConfig
from langchain.text_splitter import TokenTextSplitter
from langchain_community.vectorstores.utils import DistanceStrategy
import re
from sentence_transformers import SentenceTransformer, util
import uuid
import json
import os
import bcrypt
from rag import ContextualRetrieval
from bson.objectid import ObjectId

# Khởi tạo lifespan của ứng dụng để tải mô hình
@asynccontextmanager
async def lifespan(app: FastAPI):
    # Initialize model
    app.state.rag_model = ContextualRetrieval()
    app.state.faiss_db = None

    # Load existing conversation history if available
    history_path = "conversation_history.json"
    if os.path.exists(history_path):
        app.state.rag_model.load_conversation_history(history_path)

    # Create initial session
    app.state.initial_session_id = app.state.rag_model.start_new_session()

    # Ensure upload directory exists
    if not os.path.exists("uploaded_files"):
        os.makedirs("uploaded_files")

    yield

    # Save conversation history before shutdown
    app.state.rag_model.save_conversation_history("conversation_history.json")

app = FastAPI(lifespan=lifespan)

# Cấu hình CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Cho phép client ở localhost:3000
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

client = MongoClient('mongodb+srv://hoangkahn123:0968269517@cluster0.8to29z8.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0')  # Thay đổi chuỗi kết nối nếu cần
db = client["userdb"]
users_collection = db["users"]

# Các model cho request và response
class User(BaseModel):
    username: str
    password: str
    session_id: Optional[str] = None

class ChatRequest(BaseModel):
    query: str
    session_id: Optional[str] = None

class ChatResponse(BaseModel):
    response: str
    session_id: str

class SessionResponse(BaseModel):
    session_id: str
    message: str

class DocumentUploadResponse(BaseModel):
    message: str
    document_id: str

@app.get("/")
def status_gpu_check() -> dict[str, str]:
    gpu_msg = "Available" if len(tf.config.list_physical_devices('GPU')) > 0 else "Unavailable"
    return {
        "status": "RAG API is running",
        "gpu": gpu_msg
    }

@app.post("/session/start", response_model=SessionResponse)
def start_new_session():
    session_id = app.state.rag_model.start_new_session()
    return {
        "session_id": session_id,
        "message": "New session started successfully"
    }

@app.post("/upload/pdf", response_model=DocumentUploadResponse)
async def upload_pdf(file: UploadFile = File(...)):
    # Kiểm tra xem file có phải là PDF không
    if not file.filename.endswith('.pdf'):
        raise HTTPException(status_code=400, detail="Only PDF files are accepted")

    # Tạo document_id độc đáo
    document_id = str(uuid.uuid4())
    file_path = f"uploaded_files/{document_id}.pdf"

    # Lưu file
    with open(file_path, "wb") as buffer:
        buffer.write(await file.read())

    try:
        # Xử lý tài liệu và tạo FAISS database
        app.state.faiss_db = app.state.rag_model.process_documents(file_path)

        return {
            "message": "PDF uploaded and processed successfully",
            "document_id": document_id
        }
    except Exception as e:
        # Xóa file nếu xử lý không thành công
        if os.path.exists(file_path):
            os.remove(file_path)
        raise HTTPException(status_code=500, detail=f"Error processing PDF: {str(e)}")

@app.post("/chat", response_model=ChatResponse)
async def chat(request: ChatRequest):
    if app.state.faiss_db is None:
        raise HTTPException(status_code=400, detail="No document has been uploaded yet. Please upload a PDF first.")

    try:
        # Use the provided session_id
        session_id = request.session_id

        # Check if session exists
        if not app.state.rag_model.load_session(session_id):
            raise HTTPException(status_code=404, detail="Session not found")

        # Process the chat request
        response_text = app.state.rag_model.contextual_qa_chat(request.query, app.state.faiss_db)

        return {
            "response": response_text,
            "session_id": session_id
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=f"Error generating response: {str(e)}")

@app.get("/sessions")
def list_sessions():
    """Liệt kê tất cả các session"""
    sessions = list(app.state.rag_model.sessions.keys())
    return {"sessions": sessions}

@app.get("/history/{session_id}")
def get_session_history(session_id: str):
    """Lấy lịch sử trò chuyện của một session cụ thể"""
    if session_id not in app.state.rag_model.sessions:
        raise HTTPException(status_code=404, detail="Session not found")

    history = []
    for conv in app.state.rag_model.sessions[session_id]:
        history.append({
            "query": conv["query"],
            "response": conv["response"]
        })

    return {"session_id": session_id, "history": history}

@app.delete("/sessions", response_model=SessionResponse)
def delete_all_sessions():
    app.state.rag_model.sessions.clear()
    app.state.rag_model.current_session_id = None
    return {
        "session_id": "all",
        "message": "All sessions deleted successfully"
    }
@app.post("/register/")
async def register_user(user: User):
    # Check if username already exists
    existing_user = users_collection.find_one({"username": user.username})
    if existing_user:
        raise HTTPException(status_code=400, detail="Username already exists")

    hashed_password = bcrypt.hashpw(user.password.encode("utf-8"), bcrypt.gensalt())

    # Create a new session for the user
    session_id = app.state.rag_model.start_new_session()

    try:
        users_collection.insert_one({
            "username": user.username,
            "password": hashed_password,
            "session_id": session_id
        })
        return {"message": "User registered successfully", "session_id": session_id}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/login/")
async def login_user(user: User):
    found_user = users_collection.find_one({"username": user.username})

    if found_user and bcrypt.checkpw(user.password.encode("utf-8"), found_user["password"]):
        # Check if the user has a session_id, if not create one
        session_id = found_user.get("session_id")
        if not session_id or session_id not in app.state.rag_model.sessions:
            # Create a new session for the user
            session_id = app.state.rag_model.start_new_session()
            users_collection.update_one(
                {"_id": found_user["_id"]},
                {"$set": {"session_id": session_id}}
            )

        # Load the user's session
        app.state.rag_model.load_session(session_id)

        # Return user info with session_id
        return {
            "username": found_user["username"],
            "id": str(found_user["_id"]),
            "session_id": session_id
        }
    else:
        raise HTTPException(status_code=401, detail="Invalid username or password")
@app.post("/logout/{user_id}")
async def logout_user(user_id: str):
    try:
        # Find the user
        found_user = users_collection.find_one({"_id": ObjectId(user_id)})
        if not found_user:
            raise HTTPException(status_code=404, detail="User not found")

        # Keep the session in the database but unload it from memory
        session_id = found_user.get("session_id")
        if session_id and session_id == app.state.rag_model.current_session_id:
            app.state.rag_model.current_session_id = None

        return {"message": "Successfully logged out"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
```

```python
%%writefile app.py
from typing import Any, List, Optional
# Thêm các import cần thiết
from fastapi import FastAPI, HTTPException, UploadFile, File, Form, Request, Response, Depends
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel, Field # Thêm Field nếu cần validation chi tiết
import torch
# import tensorflow as tf # Bỏ import này nếu không dùng TF để check GPU
import os
import uuid
from contextlib import asynccontextmanager
from pymongo import MongoClient
from pymongo.errors import ConnectionFailure, OperationFailure # Thêm lỗi pymongo
import bcrypt
from rag import ContextualRetrieval # Đảm bảo file rag.py cùng thư mục hoặc trong PYTHONPATH
from bson.objectid import ObjectId # Thêm import này
import json # Thêm import này nếu chưa có
import logging # Sử dụng logging chuẩn

# Cấu hình logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# Khởi tạo lifespan của ứng dụng để tải mô hình
@asynccontextmanager
async def lifespan(app: FastAPI):
    logger.info("Starting application lifespan...")
    # Initialize model
    try:
        app.state.rag_model = ContextualRetrieval()
        app.state.faiss_db = None # FAISS DB sẽ được tạo khi upload file
        logger.info("RAG model initialized.")
    except Exception as e:
        logger.exception("Failed to initialize RAG model during startup!")
        # Có thể raise lỗi ở đây để ngăn server khởi động không thành công
        raise RuntimeError("RAG model initialization failed") from e


    # Load existing conversation history if available
    history_path = "conversation_history.json"
    if os.path.exists(history_path):
        try:
            loaded = app.state.rag_model.load_conversation_history(history_path)
            if loaded:
                logger.info(f"Loaded conversation history from {history_path}")
            else:
                 logger.warning(f"Failed to load conversation history from {history_path}, but file exists.")
        except Exception as e:
            logger.exception(f"Error loading conversation history from {history_path}")
    else:
         logger.info(f"Conversation history file not found at {history_path}, starting fresh.")

    # Ensure upload directory exists
    upload_dir = "uploaded_files"
    if not os.path.exists(upload_dir):
        try:
            os.makedirs(upload_dir)
            logger.info(f"Created directory: {upload_dir}")
        except OSError as e:
             logger.error(f"Failed to create upload directory {upload_dir}: {e}")
             # Có thể raise lỗi nếu thư mục upload là bắt buộc

    # --- Kết nối MongoDB ---
    MONGO_URI = os.getenv("MONGO_URI", "mongodb+srv://hoangkahn123:0968269517@cluster0.8to29z8.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0") # Lấy từ env var hoặc dùng default
    app.state.mongo_client = None
    app.state.users_collection = None
    try:
        logger.info(f"Connecting to MongoDB at {MONGO_URI.split('@')[-1].split('/')[0]}...") # Che password
        app.state.mongo_client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000) # Thêm timeout
        # Kiểm tra kết nối
        app.state.mongo_client.admin.command('ping')
        db = app.state.mongo_client["userdb"] # Tên database
        app.state.users_collection = db["users"] # Tên collection
        logger.info("Successfully connected to MongoDB.")
    except ConnectionFailure as e:
        logger.error(f"Failed to connect to MongoDB: {e}")
        # Server vẫn có thể chạy nhưng các chức năng liên quan DB sẽ lỗi
    except Exception as e:
        logger.error(f"An unexpected error occurred during MongoDB connection: {e}")
    # -----------------------

    yield # Application is ready to serve requests

    # --- Shutdown ---
    logger.info("Shutting down application lifespan...")
    # Save conversation history before shutdown
    if hasattr(app.state, 'rag_model'):
        try:
            app.state.rag_model.save_conversation_history("conversation_history.json")
            logger.info("Saved conversation history.")
        except Exception as e:
             logger.exception("Error saving conversation history during shutdown")

    # Close MongoDB connection
    if app.state.mongo_client:
        try:
            app.state.mongo_client.close()
            logger.info("MongoDB connection closed.")
        except Exception as e:
             logger.error(f"Error closing MongoDB connection: {e}")
    # ----------------

app = FastAPI(lifespan=lifespan)


app.add_middleware(
    CORSMiddleware,
    allow_origins=["http://localhost:5173"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*", "Authorization"],  # Thêm Authorization header
)


# Các model Pydantic
class UserCredentials(BaseModel):
    username: str
    password: str

class LoginResponse(BaseModel):
    username: str
    id: str
    session_id: str # Token sẽ được trả về trong response

class RegisterResponse(BaseModel):
    message: str
    session_id: str

class ChatRequest(BaseModel):
    query: str

class ChatResponse(BaseModel):
    response: str
    session_id: str # Trả về session đang dùng

class SessionResponse(BaseModel):
    session_id: str
    message: str

class DocumentUploadResponse(BaseModel):
    message: str
    document_id: str

class StatusResponse(BaseModel):
    status: str
    gpu: str

class SessionListResponse(BaseModel):
    sessions: List[str]

class HistoryItem(BaseModel):
     query: Optional[str] = None
     response: Optional[str] = None
     timestamp: Optional[str] = None

class HistoryResponse(BaseModel):
    session_id: str
    history: List[HistoryItem] # Sử dụng model con

class LogoutResponse(BaseModel):
     message: str


def get_users_collection():
     if not hasattr(app.state, 'users_collection') or app.state.users_collection is None:
          logger.error("Users collection is not available.")
          raise HTTPException(status_code=503, detail="Database service unavailable")
     return app.state.users_collection


def get_rag_model() -> ContextualRetrieval:
     if not hasattr(app.state, 'rag_model') or app.state.rag_model is None:
          logger.error("RAG model is not available.")
          raise HTTPException(status_code=503, detail="RAG service unavailable")
     return app.state.rag_model


# Hàm phụ thuộc để lấy session từ token
async def get_session_from_token(request: Request):
    """Extract and validate session token from Authorization header."""
    auth_header = request.headers.get("Authorization")
    if not auth_header or not auth_header.startswith("Bearer "):
        logger.warning("Access attempt failed: No valid Authorization header found.")
        raise HTTPException(status_code=401, detail="Unauthorized: Invalid or missing token")

    session_id = auth_header.split(" ")[1]

    # Có thể thêm validation nếu cần
    rag_model = get_rag_model()
    if not rag_model.load_session(session_id):
        logger.warning(f"Access attempt failed: Session ID '{session_id}' not found in RAG model memory.")
        raise HTTPException(status_code=401, detail="Invalid or expired session. Please login again.")

    return session_id


@app.get("/", response_model=StatusResponse)
def status_gpu_check():
    """Kiểm tra trạng thái API và GPU."""
    rag_model = get_rag_model() # Check if RAG model loaded
    # Sử dụng torch để kiểm tra GPU cho nhất quán
    gpu_available = torch.cuda.is_available()
    gpu_msg = f"Available ({torch.cuda.get_device_name(0)})" if gpu_available else "Unavailable"
    return {
        "status": "RAG API is running",
        "gpu": gpu_msg
    }


@app.post("/upload/pdf", response_model=DocumentUploadResponse)
async def upload_pdf(file: UploadFile = File(...)):
    """Upload và xử lý file PDF để tạo FAISS index."""
    rag_model = get_rag_model()
    if not file.filename.endswith('.pdf'):
        logger.warning(f"Upload rejected: Non-PDF file '{file.filename}'")
        raise HTTPException(status_code=400, detail="Only PDF files are accepted")

    document_id = str(uuid.uuid4())
    file_path = f"uploaded_files/{document_id}.pdf"
    logger.info(f"Received PDF upload: '{file.filename}', saving as {document_id}.pdf")

    try:
        # Lưu file an toàn hơn
        file_content = await file.read()
        if not file_content:
             raise HTTPException(status_code=400, detail="Uploaded file is empty")
        with open(file_path, "wb") as buffer:
            buffer.write(file_content)
        logger.info(f"Saved uploaded file to: {file_path}")

        # Xử lý tài liệu và tạo FAISS database
        logger.info(f"Processing PDF document: {document_id}.pdf")
        # Giả sử process_documents xử lý lỗi nội bộ và raise nếu cần
        app.state.faiss_db = rag_model.process_documents(file_path)
        logger.info(f"PDF processed successfully. FAISS DB updated for document {document_id}.")

        return {
            "message": "PDF uploaded and processed successfully",
            "document_id": document_id # Trả về ID để tham chiếu nếu cần
        }
    except FileNotFoundError as fnf_err:
         logger.error(f"Error during PDF processing: {fnf_err}")
         raise HTTPException(status_code=500, detail=f"Server error: Could not find file during processing.")
    except ValueError as val_err: # Bắt lỗi từ RAG class (vd: empty docs)
         logger.error(f"Value error during PDF processing: {val_err}")
         raise HTTPException(status_code=400, detail=str(val_err))
    except Exception as e:
        logger.exception(f"Unexpected error processing PDF {document_id}.pdf") # Ghi log cả traceback
        # Xóa file nếu xử lý không thành công
        if os.path.exists(file_path):
            try:
                os.remove(file_path)
                logger.info(f"Removed temporary file due to error: {file_path}")
            except OSError as rm_err:
                logger.error(f"Error removing file {file_path} after error: {rm_err}")
        raise HTTPException(status_code=500, detail=f"Error processing PDF: {str(e)}")
    finally:
         await file.close() # Đảm bảo file được đóng

@app.post("/chat", response_model=ChatResponse)
async def chat(request_body: ChatRequest, session_id: str = Depends(get_session_from_token)):
    """Xử lý yêu cầu chat, sử dụng session từ Authorization header."""
    rag_model = get_rag_model()

    logger.info(f"Received chat request for session: {session_id} | Query: '{request_body.query[:50]}...'")

    if app.state.faiss_db is None:
        logger.warning(f"Chat attempt failed for session {session_id}: FAISS DB not initialized.")
        raise HTTPException(status_code=400, detail="No document has been uploaded and processed yet. Please upload a PDF first.")

    try:
        # Session already loaded from the dependency function

        # Process the chat request
        logger.info(f"Processing query for session {session_id}...")
        # Giả sử contextual_qa_chat xử lý lỗi nội bộ và raise nếu cần
        response_text = rag_model.contextual_qa_chat(request_body.query, app.state.faiss_db)
        logger.info(f"Generated response for session {session_id}")

        # Trả về session_id đang dùng
        return {
            "response": response_text,
            "session_id": session_id
        }
    except HTTPException as http_exc: # Bắt lại các lỗi HTTP đã raise từ RAG hoặc check session
        raise http_exc
    except ValueError as ve: # Bắt lỗi Value Error từ RAG (vd: FAISS not ready)
         logger.error(f"Value error during chat for session {session_id}: {ve}")
         raise HTTPException(status_code=500, detail=str(ve))
    except RuntimeError as re: # Bắt lỗi Runtime từ RAG (vd: search/inference)
         logger.error(f"Runtime error during chat for session {session_id}: {re}")
         raise HTTPException(status_code=500, detail=str(re))
    except Exception as e:
        logger.exception(f"Unexpected error during chat processing for session {session_id}")
        raise HTTPException(status_code=500, detail="Internal server error during chat processing")

@app.get("/sessions", response_model=SessionListResponse)
def list_sessions(session_id: str = Depends(get_session_from_token)):
    """Liệt kê tất cả các session IDs đang có trong RAG model memory."""
    rag_model = get_rag_model()
    sessions = list(rag_model.sessions.keys())
    return {"sessions": sessions}

@app.get("/history", response_model=HistoryResponse)
def get_current_session_history(session_id: str = Depends(get_session_from_token)):
    """Lấy lịch sử trò chuyện của session hiện tại từ RAG model memory."""
    rag_model = get_rag_model()

    # Lấy lịch sử, đảm bảo trả về cấu trúc đúng như HistoryItem
    history = []
    for conv in rag_model.sessions[session_id]:
        history.append(HistoryItem(
            query=conv.get("query"),
            response=conv.get("response"),
            timestamp=conv.get("timestamp")
        ))

    return {"session_id": session_id, "history": history}

@app.get("/history/{target_session_id}", response_model=HistoryResponse)
def get_specific_session_history(
    target_session_id: str,
    current_session_id: str = Depends(get_session_from_token)
):
    """Lấy lịch sử trò chuyện của một session cụ thể từ RAG model memory."""
    rag_model = get_rag_model()
    if target_session_id not in rag_model.sessions:
        raise HTTPException(status_code=404, detail="Session not found in RAG model memory")

    # Lấy lịch sử, đảm bảo trả về cấu trúc đúng như HistoryItem
    history = []
    for conv in rag_model.sessions[target_session_id]:
        history.append(HistoryItem(
            query=conv.get("query"),
            response=conv.get("response"),
            timestamp=conv.get("timestamp")
        ))

    return {"session_id": target_session_id, "history": history}

@app.delete("/sessions", response_model=SessionResponse)
def delete_all_sessions(session_id: str = Depends(get_session_from_token)):
    """Xóa tất cả các session khỏi RAG model memory (Cẩn thận khi dùng)."""
    rag_model = get_rag_model()
    count = len(rag_model.sessions)
    rag_model.sessions.clear()
    rag_model.current_session_id = None # Reset cả session đang active
    logger.warning(f"Cleared all {count} sessions from RAG model memory.")
    return {
        "session_id": "all",
        "message": f"All {count} sessions deleted successfully from memory"
    }

@app.post("/register/", response_model=RegisterResponse, status_code=201) # Trả về 201 Created
async def register_user(user: UserCredentials): # Sử dụng UserCredentials
    """Đăng ký người dùng mới."""
    users_collection = get_users_collection()
    rag_model = get_rag_model()

    # Check if username already exists (case-insensitive check có thể tốt hơn)
    existing_user = users_collection.find_one({"username": {"$regex": f"^{user.username}$", "$options": "i"}})
    if existing_user:
        logger.warning(f"Registration failed: Username '{user.username}' already exists.")
        raise HTTPException(status_code=400, detail="Username already exists")

    try:
        hashed_password = bcrypt.hashpw(user.password.encode("utf-8"), bcrypt.gensalt())
    except Exception as e:
         logger.error(f"Error hashing password for {user.username}: {e}")
         raise HTTPException(status_code=500, detail="Error processing password")


    # Create a new RAG session for the user
    session_id = rag_model.start_new_session()
    logger.info(f"Started new RAG session {session_id} for new user {user.username}")

    try:
        user_data = {
            "username": user.username,
            "password": hashed_password,
            "session_id": session_id # Lưu session_id liên kết với user
        }
        insert_result = users_collection.insert_one(user_data)
        logger.info(f"Registered user {user.username} with DB ID {insert_result.inserted_id} and RAG session {session_id}")
        return RegisterResponse(message="User registered successfully", session_id=session_id)
    except OperationFailure as e: # Bắt lỗi cụ thể từ MongoDB
        logger.error(f"Database error during registration for {user.username}: {e.details}")
        # Cần rollback việc tạo session trong rag_model không?
        # Có thể xóa session vừa tạo: del rag_model.sessions[session_id]
        raise HTTPException(status_code=500, detail=f"Database error during registration")
    except Exception as e:
        logger.exception(f"Unexpected error registering user {user.username}")
        raise HTTPException(status_code=500, detail="Internal server error during registration")

@app.post("/login/", response_model=LoginResponse)
async def login_user(user: UserCredentials):
    """Đăng nhập người dùng và trả về token."""
    users_collection = get_users_collection()
    rag_model = get_rag_model()
    logger.info(f"Login attempt for user: {user.username}")

    # Tìm user (case-insensitive có thể tốt hơn)
    found_user = users_collection.find_one({"username": {"$regex": f"^{user.username}$", "$options": "i"}})

    if found_user and bcrypt.checkpw(user.password.encode("utf-8"), found_user["password"]):
        user_id_str = str(found_user["_id"])
        # Lấy session_id đã lưu trong DB
        session_id = found_user.get("session_id")
        session_loaded_or_created = False

        if session_id and rag_model.load_session(session_id):
             logger.info(f"Loaded existing RAG session {session_id} for user {user.username}")
             session_loaded_or_created = True
        else:
            if session_id:
                 logger.warning(f"Session ID {session_id} for user {user.username} found in DB but not in RAG model memory. Starting new session.")
            else:
                 logger.warning(f"No session ID found in DB for user {user.username}. Starting new session.")

            # Tạo session mới và cập nhật DB
            session_id = rag_model.start_new_session() # start_new_session đã set current_session_id
            try:
                update_result = users_collection.update_one(
                    {"_id": found_user["_id"]},
                    {"$set": {"session_id": session_id}}
                )
                if update_result.modified_count == 1:
                    logger.info(f"Updated user {user.username} in DB with new RAG session ID {session_id}")
                    session_loaded_or_created = True
                else:
                     logger.error(f"Failed to update session ID in DB for user {user.username}")
                     raise HTTPException(status_code=500, detail="Failed to update session in database")

            except OperationFailure as e:
                 logger.error(f"Database error updating session ID for user {user.username}: {e.details}")
                 raise HTTPException(status_code=500, detail="Database error during session update")
            except Exception as e:
                 logger.exception(f"Unexpected error updating session ID for {user.username}")
                 raise HTTPException(status_code=500, detail="Internal server error during session update")

        # Trả về thông tin kèm token nếu session được load hoặc tạo thành công
        if session_loaded_or_created:
            logger.info(f"Login successful for {user.username}. Returning session token {session_id}")

            # Trả về thông tin user và session_id như token
            return LoginResponse(
                username=found_user["username"],
                id=user_id_str,
                session_id=session_id  # Token sẽ được frontend lưu trữ
            )
        else:
             # Trường hợp này không nên xảy ra nếu logic trên đúng, nhưng để phòng ngừa
             logger.error(f"Login failed for {user.username} despite credentials being correct - session could not be loaded or created.")
             raise HTTPException(status_code=500, detail="Internal server error during session handling")

    else:
        logger.warning(f"Login failed for user {user.username}: Invalid username or password.")
        raise HTTPException(status_code=401, detail="Invalid username or password")

@app.post("/logout/{user_id}", response_model=LogoutResponse)
async def logout_user(user_id: str, session_id: str = Depends(get_session_from_token)):
    """Đăng xuất người dùng."""
    users_collection = get_users_collection()
    rag_model = get_rag_model()
    logger.info(f"Logout request for user ID: {user_id}")

    try:
        # Kiểm tra xem ObjectId có hợp lệ không trước khi truy vấn
        if not ObjectId.is_valid(user_id):
             logger.warning(f"Logout failed: Invalid user ID format '{user_id}'")
             raise HTTPException(status_code=400, detail="Invalid user ID format")

        # Tìm user không bắt buộc, nhưng tốt để log username nếu có
        found_user = users_collection.find_one({"_id": ObjectId(user_id)})
        if not found_user:
            logger.warning(f"Logout attempt for non-existent user ID: {user_id}")
            # Vẫn trả về thành công
            return LogoutResponse(message="Logout successful (user not found)")

        stored_session_id = found_user.get("session_id")
        username = found_user.get("username", "N/A")

        logger.info(f"User {username} (session {session_id}) logging out.")

        # Có thể thêm logic nếu cần (vd: lưu lịch sử, đánh dấu session)

        return LogoutResponse(message="Successfully logged out")
    except HTTPException as http_exc: # Bắt lại lỗi 400 nếu user_id không hợp lệ
        raise http_exc
    except Exception as e:
        logger.exception(f"Error during logout processing for user ID {user_id}")
        # Vẫn trả về thành công nhưng log lỗi
        return LogoutResponse(message="Logout processed with server error")
```
